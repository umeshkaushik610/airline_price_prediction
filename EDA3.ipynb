```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Merging both train and test data, as the EDA will be applied to both the datasets

### Date,month,year from DATE OF JOURNEY

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
final_df.head()
```

### Dealing with the Arrival time, we'll divide it into hour and minutes seperately

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hr'] = final_df['Arrival_Time'].str.split(':').str[0]
```

```python
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_hr'] = final_df['Arrival_hr'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

### Same goes with the Departure time

```python
final_df['Departure_hr'] = final_df['Dep_Time'].str.split(':').str[0]
```

```python
final_df['Departure_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Departure_min'] = final_df['Departure_min'].astype(int)
```

```python
final_df['Departure_hr'] = final_df['Departure_hr'].astype(int)
```

```python
final_df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

### As the Total_Stops contains just 5 types of values, so we can easily apply encoding

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head()
```

### If we observe the ROUTE feature, we find some uncertainities, to think of it in better way, we just dealt with total_stops to get an idea how many stops took in whole duration , so, here we can easily dop ROUTE as it is not necessary 

```python
final_df.drop(['Route'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

### Dealing with Duration, it is complicated - as here *hr* and *min* are used, so below is the extraction of hour and minute from duration

```python
final_df['Duration_hr'] = final_df['Duration'].str.split('h').str[0]
```

```python
# final_df[final_df['Duration_hr']=='5m']
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hr'] = final_df['Duration_hr'].astype(int)
```

```python
final_df['Duration_hr'] = final_df['Duration_hr']*60
```

```python
# final_df['Duration_mins'] = final_df['Duration'].str.split('m').str[-4:-1:1]
# final_df.head()
final_df['Duration_mins'] = final_df['Duration'].str.replace('h ',':')
```

```python
# final_df['Duration_mins'].head()
# final_df['Duration'] = final_df['Duration'].str.replace('m',' ')
```

```python
final_df['Duration_mins'].head()
```

```python
final_df['Duration_mins'] = final_df['Duration_mins'].str.split(':').str[1].str.split('m').str[0]
```

```python
# final_df['Duration_mins'] = final_df['Duration_mins'].astype(int)
# final_df[final_df['Duration_mins'] == 'NaN']
final_df['Duration_mins'] = final_df['Duration_mins'].fillna(0)
```

```python
final_df['Duration_mins'] = final_df['Duration_mins'].astype(int)
```

```python
final_df.head()
```

```python
final_df['Duration'] = final_df['Duration_hr'] + final_df['Duration_mins']
```

```python
final_df.head()
```

```python
final_df.drop(['Duration_hr','Duration_mins'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

### The airline feature contains various airlines, and as they are limited, we can easily apply LABEL ENCOING and convert them into numerical values

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
```

```python
final_df['Source'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Airline'])
```

### Now, this was all about EDA and Feature engineering, we now proceed towards applying model

```python
from sklearn.model_selection import train_test_split
```

```python
ouput = refine_train_data['Price']
```

```python
inputs = refine_train_data.drop(['Price'],axis=1)
```

```python
inputs.head()
```

```python
ouput.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,ouput,test_size=0.3)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

### We somehow got some missing values, so we dealth with them using mean

```python
# model.fit(x_train,y_train)
x_train = x_train.fillna(x_train.mean())
```

```python
model.fit(x_train,y_train)
```

```python
x_test = x_test.fillna(x_test.mean())
```

```python
y_train = y_train.fillna(y_train.mean())
```

```python
y_test = y_test.fillna(y_test.mean())
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
```
